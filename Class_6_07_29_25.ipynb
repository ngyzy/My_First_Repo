```python
#Load CSV file
import pandas as pd
import io
df = pd.read_csv("train.csv")
df
```

```python
#Correlation Heatmap - helps to check how data correlates
import seaborn as sns
import matplotlib.pyplot as plt

#Plot heatmap of numerical correlation
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap="YlGn")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Detailed Interpretation
# Pair	Correlation	Meaning
# Row ID vs Row ID	1.00	Perfect correlation (same variable)
# Row ID vs Postal Code	0.014	Very weak positive correlation
# Row ID vs Sales	0.0012	Almost no correlation
# Postal Code vs Postal Code	1.00	Perfect correlation
# Postal Code vs Sales	-0.024	Very weak negative correlation
# Sales vs Sales	1.00	Perfect correlation
```

```python
#Discount vs Sales Regression Plot
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
sns.regplot(data=df, x='Row ID', y='Sales')
plt.title('Row ID vs Sales')
plt.ylabel("Sales")
plt.xlabel("Row ID")
plt.show()
```

```python
#Sales Trend Overtime
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#Convert Order Date to Datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')

#Sort Data by Date
df=df.sort_values('Order Date')

#Plot Sales Trend Overtime
plt.figure(figsize=(12, 5))
sns.lineplot(data=df, x='Order Date', y='Sales')
plt.title('Sales Trend Over Time')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```
