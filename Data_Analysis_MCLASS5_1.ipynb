**Data Mining Association Rule**

```python
#Association rule: helps to find relationship wiithin a data.
!pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
#note apriori is a sub package within mlxtend and it's within frequent_patterns
```

```python
#Using feature engineering to add Income_Category column
df['Income_Category']=pd.cut(df['Annual Income (k$)'],bins=[0,40,70,150],labels=['Low','Medium','High'])
```

```python
df
```

```python
#Using feature engineering to add Spending_Category column
df['Spending_Category']=pd.cut(df['Spending Score (1-100)'],bins=[0,40,70,100],labels=['Low','Medium','High'])
#
```

```python
df
```

```python
#List values in variable in a structured format
transactions=df[['Gender','Income_Category','Spending_Category']].values.tolist()
```

```python
transactions
```

**ONE HOT ENCODING**

```python
#Convert textual data into zeros and 1
te=TransactionEncoder()
te_array=te.fit(transactions).transform(transactions)
df_encoded=pd.DataFrame(te_array,columns=te.columns_)
```

```python
df_encoded_with_customer_id=pd.concat([df['CustomerID'],df_encoded],axis=1)
df_encoded_with_customer_id
```

```python
#Using Frequent Item Set
#Finds item combination that appears atleast 20% of the dat
frequent_itemsets=apriori(df_encoded,min_support=0.2,use_colnames=True)
frequent_itemsets
#pass CustomerID to frequent_itemset
frequent_itemsets['length']=frequent_itemsets['itemsets'].apply(lambda x:len(x))
frequent_itemsets
```

```python
#Concat CustomerID to frequent_itemsets
frequent_itemsets_with_customer_id=pd.concat([df['CustomerID'],frequent_itemsets],axis=1)
frequent_itemsets_with_customer_id
```

```python
len(frequent_itemsets)
```

```python
#Association Rules
rules=association_rules(frequent_itemsets,metric='confidence',min_threshold=0.5)
rules
```
